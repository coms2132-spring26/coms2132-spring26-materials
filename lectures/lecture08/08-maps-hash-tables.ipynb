## 2. The Map Abstract Data Type

Map Abstract Data Type (ADT) should support all the behaviors that Python's built-in dict class supports. The most significant five methods/behaviors are:

* **get(key)**: return the value associated with a given key in a map if one exists or raise a KeyError if such value does not exist
* **set(key, value)**:  set the value for a given key to a given value. If a given key is already in a map, replace it's current value with the given value
* **remove(key)**: remove the key and its associated value. 
* **len**: return the number of key/value pairs in a map

Additionally, a map should support behaviors such as iterating through the keys, values, key-value pairs, but we will focus on the essentials. 

Note: The map ADT does not specify any requirements on the keys or values -- but specific data structure implementations may. 

Python already defines an abstract based class for the map ADT: 

````python
from collections.abc import Mapping, MutableMapping

d = {1:'a', 2:'b', 3:'c'}
isinstance(d, Mapping)
````

````python
isinstance(d, MutableMapping)
````

Python actually divides this into two classes like this: 

```
        +-------------------------------+
        | collections.abc.Mapping       |
        |-------------------------------|
        | __getitem__(self,key)         |
        | __iter__                      |
        | __len__                       |
        +-------------------------------+
                  ^
                  |  inherits
                  |
        +-------------------------------+
        | collections.abc.MutableMapping|
        |-------------------------------|
        | __setitem__                   |
        | __delitem__                   |
        | ...                           |
        +-------------------------------+
```
Note: both MutableMap and Mapping are abstract base classes they specify required methods
for the ADT, but do not implement them. But because of inheritance, any implementation for `MutableMap` should also implement the methods of `Mapping`. 

## 3. Implementing Maps

How could we implement a map using Data Structures we already know? 

### 3.1 Implementing a Map using an Array

        +-------------------------------+
        | collections.abc.MutableMapping|
        +-------------------------------+
                  ^
                  |  inherits/implements
                  |
        +-------------------------------+
        | SimpleArrayMap                |
        +-------------------------------+       

````python
class SimpleArrayMap(MutableMapping):
    
    class _Item:
        def __init__(self, k, v):
            self._key = k
            self._value = v
            
        def __eq__(self, other):
            return self._key == other._key
            

    def __init__(self): 
        self._table = [] # List of _Item instances
    
    def __getitem__(self, k):
        for item in self._table:
            if k == item._key:
                return item._value
        raise KeyError('Key not found: '+ repr(k))

    def __setitem__(self, k, v):
        for item in self._table:
            if k == item._key:
                item._value = v
                return
        self._table.append(self._Item(k, v))

    def __delitem__(self, k):
        for j in range(len(self._table)):
            if k == self._table[j]._key:
                self._table.pop(j)
                return
        raise KeyError('Key not found: '+ repr(k))

    def __len__(self):
        return len(self._table)

    def __iter__(self):
        for item in self._table:
            yield item._key
````

````python
mapping = SimpleArrayMap()

mapping["A"] = 1
mapping["B"] = 2
mapping["A"]
````

Since `__getitem__`, `__setitem__`, and `__delitem__` iterate over the list, they all have runtimes of O(n). Ideally, we want our essential data structures to have elementary operations. Can we do better?

### 3.2 TreeMap

Another option is to store the elements into a Binary Search Tree. This gives us an **ordered map** -- the keys need to be comparable to each other so that we can insert them into the BST.

```
        +-------------------------------+
        | collections.abc.MutableMapping|
        +-------------------------------+
              ^                      ^
              |                      | inherits/implements
              |                      |
  +----------------------+     +-----------+ composition  +------------------+
  | SimpleArrayMap       |     | TreeMap   |------------<>| BinarySearchTree |
  +----------------------+     +-----------+              +------------------+

````python
from collections.abc import Mapping, MutableMapping 

from binary_search_tree import BinarySearchTree

class TreeMap(MutableMapping):
    
    class _Item:
        def __init__(self, k, v):
            self._key = k
            self._value = v
            
        def __eq__(self, other):
            return self._key == other._key

        def __lt__(self, other): # make the items comparable by their key!
            return self._key < other._key

        def __gt__(self, other):
            return self._key > other._key

        def __repr__(self): 
            return f"{self._key}:{self._value}"
            
    def __init__(self): 
        self._tree = BinarySearchTree()
    
    def __getitem__(self, k):
        item = self._tree.find(self._Item(k,None)) # for lookup
        if item is None:
            raise KeyError('Key not found: '+ repr(k))
        return item._value

    def __setitem__(self, k, v):
        item = self._tree.find(self._Item(k,v)) # for lookup
        if item: # is not None
            item._value = value
        else: # otherwise insert new Item
            self._tree.insert(self._Item(k,v))    

    def __delitem__(self, k):
        raise NotImplementedError # we didn't discuss how to delete from a BST

    def __len__(self):
        return len(self._tree)

    def __iter__(self):
        for item in self._table:
            yield item._key
````

## 4. Hash Tables

* Python's implementation of **dict** class uses **hash table** data structure.
* Let's consider a specific case of a map in which all keys are integers between 0 and 10. Then, intuitively, lookup operation in a map using a key becomes similar to a lookup operation in a **lookup table** (implemented using an array) of size 11 using indices between 0 and 10 (an example of a lookup table is given below). In such table, `__getitem__`, `__setitem__` and `__delitem__` can all be implemented in O(1).

<img src="https://janakj.org/w3132/images/10.3.png" width=700/>

* But this approach is unpractical for general maps due to the following issues:
  1. Keys in maps are not necessarily integers (e.g. strings and other objects), while indices in a list have to be integers.
  2. Even if keys are integers, a given integer key might be larger than the size of the array.
  3. Using a fixed size array might be wasteful in some cases, since we might have instances when the size of the array is much bigger than the number of indices/keys and the keys with associated values may be spread out (as in the array above). 
  

### Hash functions
* To solve the issue (1), we can use **hash functions** which map keys (of any valid type) to indices in a lookup table.

<img src="https://janakj.org/w3132/images/Hash_table_3_1_1_0_1_0_0_SP.svg" width=500/>

* A **hash function** $h$ maps keys of a given type to integers in a fixed interval $[0, N − 1]$. A typical hash function for integers might be $h(x) = x mod N$. 
* The integer $h(x)$ is called the **hash value** of key $x$.
* A **hash table** for a given key type consists of hash function $h$ and an array (called table) of size $N$.
* When implementing a map with a hash table, the goal is to store item $(k, a)$ at index $i = h(k)$

One case that we need to account for is when the hash function maps more than one key to the same index. This is called a "collision". 

We will consider different options for avoiding collisions, but you can conceptualize the table as a **bucket array** which might contain multiple key-value pairs in each bucket depending on how many keys the hash function sends to the index that corresponds to each bucket.

<img src="https://janakj.org/w3132/images/10.4.png" width=700/>

#### Computing Hash Values

Computing the hash value consists of two phases: 

* **hash code**: maps a key k to an integer 
    * keys &rarr; integers
    * Typically implemented by the key objects
* **compression function**: maps the resulting integer from the code phase to an integer corresponding to an index in a bucket array
    * integers &rarr; [0, N-1]
    * Typically implemented by the hash table

<img src="https://janakj.org/w3132/images/10.5.png" width=600/>


#### Common Compression Function
* **Modulo Division**:
    * maps integer $i$ to $i$ mod $N$  (i % N), where $N$ is the size of the bucket array
    * ideally, we would take N to be a prime number to spread out the distribution of hash values, i.e., indices assigned to different hash values.
 

Let's create a basic hash table.

````python
import math

# Helper functions to ensure we keep the table size prime
def is_prime(x):
        if x < 2:
            return False
        if x % 2 == 0:
            return x == 2
        r = int(math.isqrt(x))
        for d in range(3, r + 1, 2):
            if x % d == 0:
                return False
        return True
    
def nextprime(j):
    """Return the smallest prime number p such that p >= j."""
    if j <= 2:
        return 2
    if j % 2 == 0:
        j += 1
    while not is_prime(j):
        j += 2
    return j

nextprime(24)
````

## Collision Resolution: Separate Chaining

Collision occurs when there are two or more keys with the same hash value and as a result, two different items are mapped to the same bucket. That's why we need to "spread out" the distribution of hash values and why we are using prime number for N in Modulo Division Compression Function.

A good hash code and compression function should minimize collisions. However, avoiding *all* collisions is impossible given the limited table size. 

A simple way of dealing with collisions: 
* **Separate chaining**: let each cell in the table point to a linked list (or some other contrainer) of entries that map there

<img src="https://janakj.org/w3132/images/10.6.png" width=600/>

````python
from collections.abc import MutableMapping

# VERSION 2

class SeparateChainingHashTable(MutableMapping): 

    class _Item:
        def __init__(self, k,v): 
            self._key = k
            self._value = v            

    
    def __init__(self, default_capacity = 11): 
        cap = nextprime(default_capacity) # physical size of the array
        self._n = 0 # logical number of keys
        self._table = cap * [ [] ] # each entry is an empty list -- a hash "bucket"                              

    def _hash_function(self, k):
        return hash(k) % len(self._table)

    def __len__(self):
        return self._n

    def __getitem__(self, k):
        j = self._hash_function(k)
        bucket = self._table[j]
        for item in bucket:    # changed!
            if item._key == k: # NOTE: this uses ==, i.e. __equals__ on the key
                return item._value    
        return KeyError(f"Key not in table {repr(k)}")

    def __setitem__(self, k, v):
        j = self._hash_function(k)
        bucket = self._table[j]
        for item in bucket: 
            if item._key == k: 
                return # key already in the bucket
        bucket.append(self._Item(k,v))
    
        self._n += 1
        if self._n > len(self._table) // 2:           # keep load factor <= 0.5
            self._resize(nextprime(len(self._table) * 2))

    def __delitem__(self, k):
        j = self._hash_function(k)
        bucket = self._table[j]
        for i in range(len(bucket)): 
            if item._key == k:
                del bucket[i]                
                return # key already in the bucket
        return KeyError(f"Key not in table {repr(k)}")    

    def _resize(self, c):            
        """Resize bucket array to capacity c and rehash all items."""
        old = self._table
        self._table = c * [None]       # then reset table to desired capacity
        self._n = 0                    # n recomputed during subsequent adds
        for item in old:
            if item:
                self[k] = v      # call __setitem__ to re-hash and re-insert the items        

    def __iter__(self): 
        for bucket in self._table:
            for item in bucket:
                yield item._key
````

### Collision Resolution: Open Addressing / Probing

**Open addressing**: if we don't want to use additional space like in Separate Chaining, we can use Open Addressing instead for which the colliding item is placed in a different cell of the table

The easiests type of open addressing is **Linear Probing**.

* Linear Probing handles collisions by placing the colliding item in the next (circularly) available table cell.    
* For example, the key of an item maps to index j in the table and the table A[j] is already occupied, then we try A[(j+1) mod N]. If this cell is also occupied, then we try A[(j+2) mod N] and so on until we find a cell with an empty bucket.
* Note that each table cell inspected is referred to as a “probe”.
* An issue with Linear Probing is that colliding items might lump/cluster together, causing future collisions to cause a longer sequence of probes. (**Question: How does that affect running time in the worst case?**)

<img src="https://janakj.org/w3132/images/10.7.png" width=600/>

A better option is **Quadratic Probing**: 

* If we try to insert an item into index j, and A[j] is already occupied, we try A[(j+1^2) mod N] = A[j+1 mod N]. If that cell is also occupied, we try A[(j+2^2) mod N] = A[(j+4) mod N], then A[(j+9) mod N], then A[(j+16) mod N] etc.

````python
from collections.abc import MutableMapping

# VERSION 3: Quadratic Probing Hash Map

class ProbingHashMap(MutableMapping): 

    class _Item:
        def __init__(self, k,v): 
            self._key = k
            self._value = v            

    
    def __init__(self, default_capacity = 11): 
        cap = nextprime(default_capacity) # physical size of the array
        self._n = 0 # logical number of keys
        self._table = cap * [None]                             

    def _hash_function(self, k):
        return hash(k) % len(self._table)

    def __len__(self):
        return self._n

    def __getitem__(self, k):   # currently broken!
        j = self._hash_function(k)    
        i = 1

        slot = (j + i**2) % len(self._table)
        
        while self._table[slot] != None:

            if self._table[slot]._key == k:
                return self._table[slot]._value
            
            i += 1
            slot =  (j + i**2) % len(self._table)
            if slot == j: 
                break
        raise KeyError(f"Key not found: {k}")
        

    def __setitem__(self, k, v):

        j = self._hash_function(k)    
        i = 1

        slot = (j + i**2) % len(self._table)
        
        while self._table[slot] != None:

            if self._table[slot]._key == k:
                self._table[slot]._value = v
                return                 
            
            i += 1
            slot =  (j + i**2) % len(self._table)
            
        self._table[slot] = self._Item(k,v)
        self._n += 1        
        if self._n > len(self._table) // 2:           # keep load factor <= 0.5
            self._resize(nextprime(len(self._table) * 2))

    def __delitem__(self, k):
        raise NotImplementedError   

    def _resize(self, c):            
        """Resize bucket array to capacity c and rehash all items."""
        old = self._table
        self._table = c * [None]       # then reset table to desired capacity
        self._n = 0                    # n recomputed during subsequent adds
        for item in old:
            if item:
                self[k] = v      # call __setitem__ to re-hash and re-insert the items        

    def __iter__(self): 
        for item in self._table:
            if item: 
                yield item._key
````

#### Efficiency

* The load factor is a matric which measures the ratio of the number of elements stored in the hash map to the total number of buckets available in the bucket array.
    * $\alpha = n/N$
* The load factor essentially shows how full the hash table is which determines the number of collisions.
    * In separate chaining, if the load factor approaches 1, the probability of collision increases which results in essentialy linear lookup operations (since we would have to search through large arrays)
    * In Open Addressing/Linear Probing, as the load factor grows, it takes longer to find an empty slot (we have to move around the array more to get to an empty slot)
* To keep the number of collisions manageable, we need to make sure the table is not too full. To do this, we resize the table to regain a specific load factor and reinsert items into this new table. This is called **rehashing**. The size of the new array is usually at least double the size of the previous array.
    * When done in a smart way, the overhead of the resize can be amortized, just like for arrays.

Using Hash Tables is the best way to implement a map. It is important to know that the core operations of these data structures run in constant time, making them very efficient. Since Python's dict class is implemented with hashing, it is extremely convenient to use it in different applications, since we can be sure that the main operations will be performed in constant time (on average). Note that since we are occasionally rehashing the table, the runtime of `__getitem__` and `__setitem__` will be amortized O(1). It is this underlying mechanism of hashing that makes this data structure so efficient.
