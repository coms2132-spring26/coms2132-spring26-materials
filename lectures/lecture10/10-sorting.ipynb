In general, when sorting a list of objects, Python uses the object's comparison method (`__lt__`, `__gt__`, `__eq__`).

````python
class Person:
    def __init__(self,name, age): 
        self.name = name
        self.age = age

    def __lt__(self, other): 
        if self.name == other.name: 
            return self.age < other.age # compare by age
            
        return self.name < other.name
        

    def __repr__(self): 
        return f"Person({self.name}, {self.age})"
````

The `sort` method for Python lists uses a sorting algorithm called Tim Sort, which combines Merge Sort and Insertion Sort (discussed below).

### Comparison Based Sorting Algorithms 

Comparison based sorting is the task of re-arranging a sequence into sorted order, using only comparisons between pairs of elements. 

Comparison based sorting algorithms differ in their running time behavior (as a function of the number of elements), storage requirement (additional storage required, vs. in-place) and stability. They may also differ in constant overhead. 

We can show that sorting is $\Omega(N \log N)$, i.e. it is theoretically impossible to find a sorting algorithm that is faster. 

#### Heap Sort

Using a heap, we can sort a sequence in $O(N \log N)$ time. 

* Insert elements into a new heap one-by-one, or `heapify` the original array. Best to use a max-heap (i.e. largest item at the root). 
* Then remove the largest item repeatedly, appending it to the end of the heap array. This way, no additional space is required!

Downside: the algorithm is not stable (why?)

````python
import heapq

a = [4,2,1,3,6]
heapq.heapify(a)
````

````python
a
````

````python
heapq.heappop(a)
````

````python
a
````

````python
heapq.heappush(a, 5)
````

````python
a
````

````python
def heapsort(a):

    heapq.heapify(a)
    result = []
    while a: 
        result.append(heapq.heappop(a))
    return result

heapsort([4,3,2,6,1,0,5])
````

#### Selection Sort

Selection Sort is a naive sorting algorithm that repeatedly selects the next smallest element from the array. Selection sort can be done in-place, by swapping the next-smallest element into the current position.

* Selection sort maintains a pointer to the current index (where the next smallest element will be written) -- call it p. Everything to the left of p is already sorted. Everything to the right (and including) p is unsorted. 
* In each iteration, find the smallest element in the unsorted partition. Then swap it into position p and increment p.

```
[5,4,9,3,2,6,1]

swap element at smallest index i into position p

p=0          i
[1,4,9,3,2,6,5]

increment p

   p=1        
[1,4,9,3,2,6,5]
```

Note: In the lecture on priority queues, we discussed how to implement a priority queue using an unsorted list.

````python
def selection_sort(li): # Total O(N^2)

    for p in range(len(li)):  # N steps

        #find min starting at p
        min_idx = p
        for i in range(p, len(li)):   # N steps 
            if li[i] < li[min_idx]:
                min_idx = i 

        # swap smallest element into position p
        tmp = li[p]
        li[p] = li[min_idx]
        li[min_idx] = tmp

        
````

````python
li = [4,3,2,1,0,9]
selection_sort(li)
li
````

**Selection Sort Evaluation:**

Running time: $O(N^2)$ (worst and best)

Space: $O(1)$

Stability?: No!

#### Insertion Sort

* Insertion sort maintains a pointer to the current index, also called p. Initially, p=1. 
* In each iteration, let i refer to the same index as p. If the element at index i is greater than i-1, then swap the two and decrement i. Continue swapping until the left neighbor is smaller. Then increment p.
* Essentially, we are inserting the element at index p into the right position to the left of p.

```
   p=1
[5,4,9,3,2,6,1]

i=p=4
swap with i-1 and decrement i

 i p=1
[4,5,9,3,2,6,1]

increment p
     p
[4,5,9,3,2,6,5]
```
[example continued in class]


Note: In the lecture on priority queues, we discussed how to implement a priority queue using a sorted list.

````python
def insertion_sort(li):
    for p in range(1, len(li)):

        i = p 
        while i > 0 and li[i] < li[i-1]:
            tmp = li[i-1]
            li[i-1] = li[i]
            li[i] = tmp
            i = i - 1          
        
````

````python
li = [4,3,2,1,0,9]
insertion_sort(li)
li
````

**Insertion Sort Evaluation:**

Running time: $O(N^2)$ worst case, $O(N)$ best case!

Space: $O(1)$

Stability?: YES!

#### Merge Sort

* Divide the unsorted list into two equal-sized partitions 
* Recursively sort each half
* Then merge the two halfs back together 

```
[5,4,9,3,2,6,1]
Split

[5,4,9] [2,6,1]
Sort recursively

[4,5,9] [1,2,6]

Merge
[1,2,4,5,6,9]
```

**The merge algorithm**: pointer i into the left array, pointer j into the right array. Compare the two elements. Write the smaller element into a new array and increment the corresponding pointer. 

````python
def merge(sorted_left, sorted_right): 
    result = []
    i = 0 
    j = 0 
    while i<len(sorted_left) and j<len(sorted_right): 
        if sorted_left[i] <= sorted_right[j]:
            result.append(sorted_left[i])
            i += 1
        else:
            result.append(sorted_right[j])
            j += 1
    while i<len(sorted_left): 
        result.append(sorted_left[i])
        i += 1
    while j<len(sorted_right): 
        result.append(sorted_right[j])
        j += 1
    return result 

def merge_sort(a): 

    if len(a) == 1: 
        return a
    
    mid = len(a) // 2
    left = a[0:mid]
    right = a[mid:]

    sorted_left = merge_sort(left)
    sorted_right = merge_sort(right)

    #merge 
    return merge(sorted_left, sorted_right)
````

````python
merge_sort([1,5,9,0,3,7,4,6])
````

**Merge Sort Evaluation:**

Running time: $O(N log N)$ even in the best case

Space: $O(N)$

Stability?: yes (because merge preserves order of duplicates)

#### Quick Sort

Quick Sort is also recursive. Identify a pivot element (typically at the mid position, but the pivot selection influences running time). 
Then **partition** the array into elements smaller than the pivot (move to the left) and elements greater than the pivot (move to the right).
Then recursively sort each of the two sub-arrays. 

````python
def quick_sort_rec(a, left, right): 

    print("Sorting",left, right, a[left:right+1])
    if left >= right: 
        return 
        
    pivot_idx = (left + right) // 2 # select pivot in the middle

    # swap pivot to end
    pivot = a[pivot_idx]
    print("Pivot: ", pivot)
    a[pivot_idx] = a[right]
    a[right] = pivot 
    print(a)
    
    # partition 
    i = left
    j = right - 1
    while i < j:    
        while a[i] < pivot:
            i += 1
        while a[j] > pivot: 
            j -= 1
        if j > i:
            #swap i and j
            print("Swap ", a[i], a[j])
            tmp = a[i]
            a[i] = a[j]
            a[j] = tmp
            print(a)
    
    # swap pivot back to position i
    a[right] = a[i]
    a[i] = pivot
    print("After partition", a)


    quick_sort_rec(a, left, i-1)
    quick_sort_rec(a, i+1, right)    
    return a

def quick_sort(a): 
    return quick_sort_rec(a, 0, len(a)-1)
````

````python
a = [7,1,4,2,3,8]
quick_sort(a)
````

Running time: $O(N^2)$ worst case  $O(N)$ best case
